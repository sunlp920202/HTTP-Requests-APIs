```python
from requests.auth import HTTPBasicAuth
import requests
import pandas as pd

#API call to the Students Talbe
URL="//api/v5/schools/1/students/"
Header={
    "AERIES-CERT":"sample"
}
Parameters={
    "DatabaseYear": 2024
}

response=requests.get(url=URL, headers=Header, json=Parameters)
response.raise_for_status()
print(response)

data=response.json()
print(data)
```

```python
Students: traditioal way 

Transcripts:append and concat way

Tests: scenario
```

```python
stu_id=[]
stu_sc=[]
stu_sn=[]
stu_cid=[]
stu_ln=[]
stu_fn=[]
stu_gn=[]
stu_gr=[]
stu_bd=[]
stu_ma=[]
stu_city=[]
stu_state=[]
stu_zip=[]
stu_eth=[]
stu_rc1=[]
stu_se=[]
stu_sl=[]
stu_de=[]
stu_fk=[]
stu_lf=[]
stu_pe=[]
stu_pemail=[]
stu_semail=[]



for i in data:
    stu_id.append(i["StudentID"])
    stu_sc.append(i["SchoolCode"])
    stu_sn.append(i["StudentNumber"])
    stu_cid.append(i["StateStudentID"])
    stu_ln.append(i["LastName"])
    stu_fn.append(i["FirstName"])
    stu_gn.append(i["Gender"])
    stu_gr.append(i["Grade"])
    stu_bd.append(i["Birthdate"])
    stu_ma.append(i["MailingAddress"])
    stu_city.append(i["MailingAddressCity"])
    stu_state.append(i["MailingAddressState"])
    stu_zip.append(i["MailingAddressZipCode"])
    stu_eth.append(i["EthnicityCode"])
    stu_rc1.append(i["RaceCode1"])
    stu_se.append(i["SchoolEnterDate"])
    stu_sl.append(i["SchoolLeaveDate"])
    stu_de.append(i["DistrictEnterDate"])
    stu_fk.append(i["FamilyKey"])
    stu_lf.append(i["LanguageFluencyCode"])
    stu_pe.append(i["ParentEdLevelCode"])
    stu_pemail.append(i["ParentEmailAddress"])
    stu_semail.append(i["StudentEmailAddress"])
    
students=pd.DataFrame({
    "StudentID":stu_id,
    "SchoolCode":stu_sc,
    "StudentNumber":stu_sn,
    "StateStudentID":stu_cid,
    "LastName":stu_ln,
    "FirstName":stu_fn,
    "Gender":stu_gn,
    "Grade":stu_gr,
    "Birthdate":stu_bd,
    "MailingAddress":stu_ma,
    "MailingAddressCity":stu_city,
    "MailingAddressState":stu_state,
    "MailingAddressZipCode":stu_zip,
    "EthnicityCode":stu_eth,
    "RaceCode1":stu_rc1,
    "SchoolEnterDate":stu_se,
    "SchoolLeaveDate":stu_sl,
    "DistrictEnterDate":stu_de,
    "FamilyKey":stu_fk,
    "LanguageFluencyCode":stu_lf,
    "ParentEdLevelCode":stu_pe,
    "ParentEmailAddress":stu_pemail,
    "StudentEmailAddress":stu_semail
})

students.head()
```

```python
#API call to the Tests
base_url="//api/v5/students/{StudentID}/tests"
students_id=stu_id[0:10]#Get the elements from student list from index 0 to 10
Header={
    "AERIES-CERT":"sample"
}
Parameters={
    "DatabaseYear": 2024
}
tests=[]
for student_id in students_id:
    URL3=base_url.replace('{StudentID}',str(student_id))
    response3=requests.get(url=URL3, headers=Header, json=Parameters)
    response3.raise_for_status()
    print(response3)

    data3=response3.json()
    test_sid=[]
    test_1=[]
    test_2=[]
    test_3=[]
    test_4=[]
    test_5=[]
    test_6=[]
    test_sq=[]
    test_tid=[]
    test_part=[]
    test_dc=[]
    test_gl=[]
    test_ta=[]
    test_td=[]

    for k in data3:
        test_sid.append(k['StudentID'])
    
        scenario=len(k['Scores'])
        if scenario==1:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["TestID"])
            test_3.append(k["TestID"])
            test_4.append(k["TestID"])
            test_5.append(k["TestID"])
            test_6.append(k["TestID"])
        elif scenario==2:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["Scores"][1]["Score"])
            test_3.append(k["TestID"])
            test_4.append(k["TestID"])
            test_5.append(k["TestID"])
            test_6.append(k["TestID"])
        elif scenario==3:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["Scores"][1]["Score"])
            test_3.append(k["Scores"][2]["Score"])
            test_4.append(k["TestID"])
            test_5.append(k["TestID"])
            test_6.append(k["TestID"])
        elif scenario==4:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["Scores"][1]["Score"])
            test_3.append(k["Scores"][2]["Score"])
            test_4.append(k["Scores"][3]["Score"])
            test_5.append(k["TestID"])
            test_6.append(k["TestID"])
        elif scenario==5:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["Scores"][1]["Score"])
            test_3.append(k["Scores"][2]["Score"])
            test_4.append(k["Scores"][3]["Score"])
            test_5.append(k["Scores"][4]["Score"])
            test_6.append(k["TestID"])
        elif scenario==6:
            test_1.append(k["Scores"][0]["Score"])
            test_2.append(k["Scores"][1]["Score"])
            test_3.append(k["Scores"][2]["Score"])
            test_4.append(k["Scores"][3]["Score"])
            test_5.append(k["Scores"][4]["Score"])
            test_6.append(k["Scores"][5]["Score"])
        else:
            test_1.append(k["TestID"])
            test_2.append(k["TestID"])
            test_3.append(k["TestID"])
            test_4.append(k["TestID"])
            test_5.append(k["TestID"])
            test_6.append(k["TestID"])
        
        test_sq.append(k["SequenceNumber"])
        test_tid.append(k["TestID"])
        test_part.append(k["TestPart"])
        test_dc.append(k["TestDescription"])
        test_gl.append(k["GradeLevel"])
        test_ta.append(k["TestingAdministration"])
        test_td.append(k["TestDate"])
 
    tests_df=pd.DataFrame({
        "StudentID":test_sid,
        "Test1":test_1,
        "Test2":test_2,
        "Test3":test_3,
        "Test4":test_4,
        "Test5":test_5,
        "Test6":test_6,
        "SequenceNumber":test_sq,
        "TestID":test_tid,
        "TestPart":test_part,
        "TestDescription":test_dc,
        "GradeLevel":test_gl,
        "TestingAdministration":test_ta,
        "TestDate": test_td
     })

    tests.append(tests_df)
#Combine the created dataframes into a list of dataframes.

#Convert the list of dataframes into one dataframe.
tests_all=pd.concat(tests, ignore_index=True)
tests_all.head(100)
```
